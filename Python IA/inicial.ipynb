### Case: Score de Crédito dos Clientes

Você foi contratado por um banco para conseguir definir o score de crédito dos clientes. Você precisa analisar todos os clientes do banco e, com base nessa análise, criar um modelo que consiga ler as informações do cliente e dizer automaticamente o score de crédito dele: Ruim, Ok, Bom


```python
#importar base de dados




import pandas as pd

table=pd.read_csv("clientes.csv")
display(table)

#preparar base de dados para a IA (pandas, scikit-learn)

display(table.info()) #mostrar infomacoes da table

#profissoes
    #mecanico 1, professor 2, artista 3, advogado 4, medico 5, bombeiro 6, policial 7

#label encoder(transf textos em numeros)
from sklearn.preprocessing import LabelEncoder

codificador=LabelEncoder()

table["profissao"]=codificador.fit_transform(table["profissao"]) #transf as profissoes em numeros



#mix_credito
codificador2=LabelEncoder()

table["mix_credito"]=codificador2.fit_transform(table["mix_credito"])

#comportamento_pagamento

codificador3=LabelEncoder()
table["comportamento_pagamento"]=codificador3.fit_transform(table["comportamento_pagamento"])

display(table.info())


```

```python
#y é quem eu quero prever
#x é quem eu vou usar pra fazer a previsao
y=table["score_credito"]
x=table.drop(columns=["score_credito","id_cliente"])



from sklearn.model_selection import train_test_split  #funcao pra dividir os dados e realizar treino

x_treino,x_test,y_treino,y_test=train_test_split(x,y,test_size=0.3)
```

```python
#criar modelo de IA, arvore de decisao(randomforest), vizinhosproximos(nearest neighbors)

#importando IA
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

#criando os modelos

modelo_arvore=RandomForestClassifier()
modelo_knn=KNeighborsClassifier()

#train AI
modelo_arvore.fit(x_treino, y_treino)
modelo_knn.fit(x_treino,y_treino)

#avaliar qual melhor modelo de IA
previsao_arvore=modelo_arvore.predict(x_test)
previsao_knn=modelo_knn.predict(x_test)


#acuracia, mostrar os acertos de cada
from sklearn.metrics import accuracy_score
display(accuracy_score(y_test, previsao_arvore))
display(accuracy_score(y_test, previsao_knn))



#foi definido que o modelo arvore é melhor
#fazer novas previsoes de acordo com novos clientes

table_newcli=pd.read_csv("novos_clientes.csv")
display(table_newcli)

table_newcli["profissao"]=codificador.fit_transform(table_newcli["profissao"]) #transf as profissoes em numeros

table_newcli["mix_credito"]=codificador2.fit_transform(table_newcli["mix_credito"])


table_newcli["comportamento_pagamento"]=codificador3.fit_transform(table_newcli["comportamento_pagamento"])

previsao=modelo_arvore.predict(table_newcli)
```
